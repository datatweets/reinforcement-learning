## 1. What is Reinforcement Learning?

Think about how you learned to ride a bicycle:
- Nobody gave you a manual with 10,000 labeled examples of "correct pedaling" (that would be supervised learning)
- Nobody asked you to find hidden patterns in bicycle data (that would be unsupervised learning)
- Instead, you **got on the bike, fell down, adjusted, and tried again**. Every time you stayed balanced a bit longer, your brain said "do more of that!" Every time you fell, your brain said "don't do that!"

**That's Reinforcement Learning** — learning by doing, through trial and error, guided by rewards and penalties.

### The Simple Definition

> **Reinforcement Learning (RL)** = An agent learns to make good decisions by interacting with an environment and receiving feedback (rewards).

### Key Characteristics

| What Makes RL Special | Example |
|---|---|
| **Learning by doing** — no teacher, no answer key | A baby learning to walk by trying and falling |
| **Delayed feedback** — rewards may come later | Studying for months before getting exam results |
| **Actions have consequences** — what you do now affects the future | Making a chess move affects the entire game |
| **Exploration needed** — you must try new things | Trying a new restaurant instead of your usual one |

## 2. The Five Building Blocks of RL

Every RL problem has exactly five components. Let's learn them through a real example.

### 🎮 Example: Teaching an AI to Play a Video Game

Imagine you're building an AI to play Super Mario:

#### 🤖 Agent — *The Learner*
The AI player. It looks at the screen, decides what to do, and learns from what happens.

#### 🌍 Environment — *The World*
The game itself — the level, the enemies, the coins, the platforms. Everything the agent interacts with.

#### 📍 State — *What's Happening Right Now*
A snapshot of the current situation: Mario's position, nearby enemies, remaining time, score.

#### ⚡ Action — *What the Agent Can Do*
The buttons Mario can press: Jump, Move Right, Move Left, Run, Duck.

#### 🏆 Reward — *The Feedback Signal*
- Collect a coin → **+1** (good!)
- Kill an enemy → **+5** (great!)  
- Fall into a pit → **-10** (bad!)
- Each second that passes → **-0.1** (hurry up!)

### The RL Loop — How It All Connects

```
     ┌──────────┐                    ┌──────────────┐
     │          │ ── "Jump Right" ──►│              │
     │  AGENT   │                    │ ENVIRONMENT  │
     │ (Mario   │ ◄── "Score +1,  ──│ (The Game)   │
     │   AI)    │     New Screen"   │              │
     └──────────┘                    └──────────────┘
     
     Repeats thousands of times per game!
     
     Step 1: Agent sees the current screen (State)
     Step 2: Agent picks a button to press (Action)
     Step 3: Game responds with new screen (New State) and points (Reward)
     Step 4: Agent learns from this experience
     Step 5: Repeat!
```

## 3. How is RL Different from Other Machine Learning?

There are three main types of machine learning. Here's the simplest way to understand the difference:

| | Supervised Learning | Unsupervised Learning | Reinforcement Learning |
|---|---|---|---|
| **Analogy** | Learning from a textbook with answer key | Finding patterns in a puzzle | Learning to ride a bike |
| **Data** | Labeled examples (X → Y) | Unlabeled data | No data upfront — generates its own! |
| **Feedback** | "Here's the right answer" | No feedback | "That was good" or "That was bad" |
| **Goal** | Predict correct labels | Find hidden structure | Maximize total reward |
| **Examples** | Spam detection, image recognition | Customer segmentation, clustering | Game AI, robotics, trading |

### The Key Differences in Plain Language

**Supervised Learning** = You have a teacher who shows you the right answer for every question. *"This email IS spam. This one is NOT spam."*

**Unsupervised Learning** = You have data but no teacher. You're looking for patterns. *"These customers seem similar to each other."*

**Reinforcement Learning** = No teacher, no answer key. You try things and get a score. *"I don't know the best move, but that last move got me 5 points."*

## 5. Let's Build Our First RL Agent! 🚀

Enough theory — let's build something! We'll create a simple world and teach an agent to navigate it.

### 5.1 The Grid World

Our agent (A) needs to reach the Goal (G) while avoiding the Trap (X). Each step costs -1 to encourage finding the shortest path.

```
┌───┬───┬───┬───┐
│   │   │   │ G │    G = Goal   → Reward: +10
├───┼───┼───┼───┤    X = Trap   → Reward: -10
│   │ X │   │   │    Each step  → Reward: -1
├───┼───┼───┼───┤
│ A │   │   │   │    A starts at bottom-left
└───┴───┴───┴───┘    Must reach G at top-right!
    
Actions: Up, Right, Down, Left
```

````python
import numpy as np
import random

# ============================================================
# Our First RL Environment: Grid World
# ============================================================
class GridWorld:
    def __init__(self):
        self.rows = 3
        self.cols = 4
        self.start = (2, 0)       # Bottom-left
        self.goal = (0, 3)        # Top-right
        self.trap = (1, 1)        # Middle area
        self.state = self.start
        
        # Actions: 0=Up, 1=Right, 2=Down, 3=Left
        self.actions = {0: (-1, 0), 1: (0, 1), 2: (1, 0), 3: (0, -1)}
        self.action_names = {0: 'Up', 1: 'Right', 2: 'Down', 3: 'Left'}
        self.n_actions = 4
    
    def reset(self):
        self.state = self.start
        return self.state
    
    def step(self, action):
        dx, dy = self.actions[action]
        new_row = max(0, min(self.rows - 1, self.state[0] + dx))
        new_col = max(0, min(self.cols - 1, self.state[1] + dy))
        self.state = (new_row, new_col)
        
        if self.state == self.goal:
            return self.state, +10.0, True     # Reached goal!
        elif self.state == self.trap:
            return self.state, -10.0, True     # Fell in trap!
        else:
            return self.state, -1.0, False     # Keep going
    
    def render(self):
        symbols = {self.goal: ' G ', self.trap: ' X ', self.state: ' A '}
        print("+----+----+----+----+")
        for r in range(self.rows):
            row_str = "|"
            for c in range(self.cols):
                pos = (r, c)
                if pos in symbols:
                    row_str += f" {symbols[pos]}|"
                else:
                    row_str += "  . |"
            print(row_str)
            print("+----+----+----+----+")

# Let's see our world!
env = GridWorld()
print("Our Grid World:")
print("Agent = A, Goal = G, Trap = X\n")
env.reset()
env.render()
print(f"\nAgent starts at: {env.start}")
print(f"Goal is at:      {env.goal}")
print(f"Trap is at:      {env.trap}")
````

### 5.2 The Random Agent — No Learning

First, let's see what happens when an agent takes **completely random actions** — no learning at all. This gives us a baseline to compare against.

````python
# ============================================================
# Random Agent: No learning, just random button mashing
# ============================================================
def run_random_agent(env, n_episodes=5000, max_steps=50):
    results = {'wins': 0, 'traps': 0, 'timeouts': 0, 'total_rewards': []}
    for episode in range(n_episodes):
        state = env.reset()
        total_reward = 0
        for step in range(max_steps):
            action = random.randint(0, 3)   # Pick random action
            state, reward, done = env.step(action)
            total_reward += reward
            if done:
                if state == env.goal:
                    results['wins'] += 1
                else:
                    results['traps'] += 1
                break
        else:
            results['timeouts'] += 1
        results['total_rewards'].append(total_reward)
    return results

env = GridWorld()
random_results = run_random_agent(env, n_episodes=5000)

print("=" * 50)
print("RANDOM AGENT RESULTS (5,000 games)")
print("=" * 50)
print(f"  Reached Goal:   {random_results['wins']:>5} ({random_results['wins']/50:.1f}%)")
print(f"  Fell in Trap:   {random_results['traps']:>5} ({random_results['traps']/50:.1f}%)")
print(f"  Ran out of time:{random_results['timeouts']:>5} ({random_results['timeouts']/50:.1f}%)")
print(f"  Average Score:  {np.mean(random_results['total_rewards']):.1f}")
print(f"\nNot great! The random agent just stumbles around.")
print(f"Can we do better? YES -- with Q-Learning!")
````

### 5.3 The Q-Learning Agent — Learning from Experience

Now let's build an agent that actually **learns**. Q-Learning works like this:

1. Keep a "cheat sheet" (Q-table) that stores: *"How good is each action in each state?"*
2. Start with no knowledge (all zeros)
3. Try actions, see what happens, and update the cheat sheet
4. Over time, the cheat sheet gets really good!

The update rule is simple:

```
New Knowledge = Old Knowledge + Learning Rate * (What Actually Happened - What I Expected)
```

````python
# ============================================================
# Q-Learning Agent: Learns from experience!
# ============================================================
def train_q_learning(env, n_episodes=2000, alpha=0.1, gamma=0.99,
                     epsilon_start=1.0, epsilon_end=0.01, epsilon_decay=0.995):
    # Initialize Q-table: stores "how good is action A in state S?"
    q_table = {}
    for r in range(env.rows):
        for c in range(env.cols):
            q_table[(r, c)] = np.zeros(env.n_actions)
    
    epsilon = epsilon_start   # Start by exploring a lot
    episode_rewards = []
    wins_per_100 = []
    wins = 0
    
    for episode in range(n_episodes):
        state = env.reset()
        total_reward = 0
        
        for step in range(100):
            # EXPLORATION vs EXPLOITATION
            if random.random() < epsilon:
                action = random.randint(0, 3)          # Explore: try random
            else:
                action = np.argmax(q_table[state])     # Exploit: use best known
            
            next_state, reward, done = env.step(action)
            total_reward += reward
            
            # THE LEARNING STEP (Q-Learning update)
            # Q(s,a) = Q(s,a) + alpha * (reward + gamma * best_future - Q(s,a))
            best_future = np.max(q_table[next_state])
            q_table[state][action] += alpha * (
                reward + gamma * best_future - q_table[state][action]
            )
            
            state = next_state
            if done:
                if state == env.goal:
                    wins += 1
                break
        
        epsilon = max(epsilon_end, epsilon * epsilon_decay)  # Explore less over time
        episode_rewards.append(total_reward)
        
        if (episode + 1) % 100 == 0:
            wins_per_100.append(wins)
            wins = 0
    
    return q_table, episode_rewards, wins_per_100

# Train the agent!
env = GridWorld()
q_table, rewards, wins_per_100 = train_q_learning(env)

print("=" * 50)
print("Q-LEARNING AGENT TRAINING COMPLETE!")
print("=" * 50)
print(f"  Episodes trained:        2,000")
print(f"  Avg reward (last 100):   {np.mean(rewards[-100:]):.1f}")
print(f"  Win rate (last 100):     {wins_per_100[-1]}%")
print(f"\nCompare to random agent:   {random_results['wins']/50:.1f}% win rate")
print(f"\nThe agent LEARNED to reach the goal!")
````

````python
# ============================================================
# Watch the Trained Agent Play!
# ============================================================
def watch_agent(env, q_table, max_steps=20):
    state = env.reset()
    print("=" * 50)
    print("WATCHING TRAINED AGENT PLAY")
    print("=" * 50)
    total_reward = 0
    path = [state]
    
    for step in range(max_steps):
        action = np.argmax(q_table[state])
        print(f"\nStep {step+1}: At {state}, Agent chooses '{env.action_names[action]}'")
        
        state, reward, done = env.step(action)
        total_reward += reward
        path.append(state)
        env.render()
        
        if done:
            if state == env.goal:
                print(f"\nGOAL REACHED in {step+1} steps! Score: {total_reward:+.1f}")
            else:
                print(f"\nTrapped! Score: {total_reward:+.1f}")
            break
    
    print(f"\nPath: {' -> '.join(str(p) for p in path)}")

watch_agent(env, q_table)
````

````python
# ============================================================
# Random Agent on CartPole — How Bad Is It?
# ============================================================
import gymnasium as gym
import numpy as np

env = gym.make('CartPole-v1')
episode_lengths = []

for episode in range(1000):
    state, info = env.reset()
    steps = 0
    while True:
        action = env.action_space.sample()    # Random!
        state, reward, terminated, truncated, info = env.step(action)
        steps += 1
        if terminated or truncated:
            break
    episode_lengths.append(steps)

env.close()

print("=" * 55)
print("RANDOM AGENT ON CARTPOLE (1,000 games)")
print("=" * 55)
print(f"  Average survival:  {np.mean(episode_lengths):.1f} / 500 steps")
print(f"  Best game:         {max(episode_lengths)} steps")
print(f"  Worst game:        {min(episode_lengths)} steps")
print(f"\n  The random agent barely keeps the pole up!")
print(f"  We'll fix this with proper RL in Parts 3 & 4.")
````
